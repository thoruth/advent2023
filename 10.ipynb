```python
data = live_input.split('\n')

dim_x, dim_y = len(data[0]), len(data)

dim_x, dim_y

def to_n(x, y):
    if x<0 or y<0:
        return None
    return x + y * dim_x

def sides(x, y, mode):
    if mode=='north':
        return (x, y-1)
    elif mode=='south':
        return (x, y+1)
    elif mode=='east':
        return (x+1, y)
    elif mode=='west':
        return (x-1, y)

def connect(x,y, pipe_type):
    if '.'==pipe_type:
        return []
    elif 'S'==pipe_type:
        return [sides(x, y, dir) for dir in ['north', 'east', 'south', 'west']]
    elif '|'==pipe_type:
        return [sides(x, y, 'north'), sides(x, y, 'south')]
    elif '-'==pipe_type:
        return [sides(x, y, 'east'), sides(x, y, 'west')]
    elif 'L'==pipe_type:
        return [sides(x, y, 'north'), sides(x, y, 'east')]
    elif 'J'==pipe_type:
        return [sides(x, y, 'north'), sides(x, y, 'west')]
    elif '7'==pipe_type:
        return [sides(x, y, 'south'), sides(x, y, 'west')]
    elif 'F'==pipe_type:
        return [sides(x, y, 'south'), sides(x, y, 'east')]
    elif '.'==pipe_type:
        return []
    

graph = {}
start = None
dot_chars = []
for idy, row in enumerate(data):
    x_row = []
    for idx, char in enumerate(row):
        n = to_n(idx, idy)
        if char == 'S':
            start = n
        elif char == '.':
            dot_chars.append((idx, idy))
        mapped = [ to_n(value[0], value[1]) for value in connect(idx, idy, char)]
        graph[n] = [x for x in mapped if x ]
```

```python

network = []
for from_, tos_ in graph.items():
    if tos_:
        for to_ in tos_:
            if to_ in graph and from_ in graph[to_]:
                
                network.append((from_, to_))
```

```python
import networkx as nx

G = nx.Graph(list(network))
p = nx.shortest_path(G, source=start)
```

```python
def draw(paths, loop=None):
    final_str = ''
    for y in range(dim_y):
        str_row=''
        for x in range(dim_x):
            n = to_n(x, y)
            
            if loop is None and n in paths:
                str_row +=data[y][x]
            elif loop is not None and n in loop:
                str_row +=data[y][x]
            else:
                str_row+='.'
        final_str +=str_row + '\n'
    print(f'{final_str}')
```
