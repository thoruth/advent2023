```python
class num_with_pos:

    def __init__(self, num, start, end, line):
        self.num=int(num)
        self.start = start
        self.end = end
        self.line = line
        self.has_gear = False
        self.gear_pos = None

    def __str__(self):
        return f'<num-{self.num} start-{self.start} end-{self.end} line-{self.line}>'

    def __repr__(self):
        return self.__str__()
```

```python
def parse_line(line, line_num):
    res = []
    regexmatches = re.finditer(r'\d+', a)
    for regexmatch in regexmatches:
        res.append(num_with_pos(regexmatch.group(), regexmatch.start(), regexmatch.end()-1, line_num))
    return res

def get_surround_check(num_obj:num_with_pos, width, height):

    start = num_obj.start
    end = num_obj.end
    line = num_obj.line
    ###
    #
    # xxx
    # x0x
    # xxx
    ## up part
    res = []
    start_x= max(start -1, 0)
    start_y = max(line-1, 0)
    end_x = min(end +1, width)
    end_y = min(line+1, height)
    res = []
    for y in range(start_y,end_y+1):
        for x in range(start_x, end_x+1):
            if y == line and x >=start and x <= end:
                continue   
            else:
                res.append((x,y))
    return res


final_nums = []
data =live_inputs.split('\n')
# data =test_inputs.split('\n')
height = len(data) -1
width = len(data[0]) -1

for idx, a in enumerate(data):
    test_pa = parse_line(a, idx)
    final_nums += test_pa

```

```python
def filter_nums(num:num_with_pos, width, height):
    sorround_check = get_surround_check(num, width, height)
    
    for char_pos in sorround_check:
        char = data[char_pos[1]][char_pos[0]]
        if char == '*':
            num.has_gear = True
            num.gear_pos = char_pos
            return True
        elif char != '.' and not char.isnumeric():
            return True
    return False
```

```python
filtered_nums = []
geared_nums =[]
for num in final_nums:
    if filter_nums(num, width, height):
        filtered_nums.append(num)
        if num.has_gear:
            geared_nums.append(num)


print(geared_nums)
print(filtered_nums)
# sum([num.num for num in filtered_nums])
```

```python
geared_res = {}
for num in geared_nums:
    key = str(num.gear_pos)
    if key in geared_res:
        value = geared_res[key]['value'] 
        if not value:
            geared_res[key]['value'] = geared_res[key]['first']*num.num
        else: 
            geared_res[key]['value'] = value*num.num

    else:
        geared_res[key] ={'first':num.num, 'value':0}
geared_res 
sum([x['value'] for x in geared_res.values()]) + sum([num.num for num in filtered_nums]) 
```
